```python
from pathlib import Path  
import cv2  
import numpy as np  
  
#  .jpg   
search_path = Path('./data')  
  
#  pathlib.Path.glob  .jpg   
jpg_files = list(search_path.glob('**/*.jpg'))  
  
#   
data = [(file.stem, cv2.cvtColor(cv2.imread(str(file)), cv2.COLOR_BGR2GRAY)) for file in jpg_files]  
  
#   
for img_name, img in data:  
    print(f"Image Name: {img_name}, Shape: {img.shape}")
```
